```python
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt

pizza = imread('pizza.jpg', as_gray=True)

plt.imshow(pizza)
plt.show()
pizza.shape
```

```python
def qualMaisProximo(listaValores, chave):
#     print(chave,listaValores)
    maisProximo = 999999999.99
    distanciamaisProximo = 999999999.99
    posicaoMaisProximo = 9999

    for i in range(len(listaValores)):
        if( abs(listaValores[i]-chave) < distanciamaisProximo):
            maisProximo = listaValores[i]
            distanciamaisProximo = abs(listaValores[i]-chave)
            posicaoMaisProximo = i
            
    return maisProximo
```

#   algoritmo da filtragem pela media utilizando janela 3 x 3.

```python

for i in range(len(pizza)):
    for j in range(len(pizza[i])):
        vizinhos =[]
        valorM1 = pizza[i][j] #         valor do pixem em pizza
        
        vizinhos.append(pizza[max(i-1,0)][max(j-1,0)])
        vizinhos.append(pizza[max(i-1,0)][j])
        vizinhos.append(pizza[max(i-1,0)][min(j+1,599)])
        vizinhos.append(pizza[i][max(j-1,0)])
        vizinhos.append(pizza[i][min(j+1,599)])
        vizinhos.append(pizza[min(i+1,399)][max(j-1,0)])
        vizinhos.append(pizza[min(i+1,399)][j])
        vizinhos.append(pizza[min(i+1,399)][min(j+1,599)])
        
        #maisProximo = qualMaisProximo(vizinhos,valorM1)
        #print("O valor mais proximo de ",valorM1,' dentro da vizinhana de 3x3 pixels  o: ', maisProximo)
        # preciso pegar os k valores mais prximos de valorM1 numa janela de i+Janela, i-Janela
        #pizza[i][j] = maisProximo

        soma = 0
        for x in range(len(vizinhos)):
            soma += vizinhos[x]
        
        
        res = soma/8

        print(pizza[i][j], " recebe ",res)

        pizza[i][j] = res
    
plt.imshow(pizza)
plt.show() 
```
