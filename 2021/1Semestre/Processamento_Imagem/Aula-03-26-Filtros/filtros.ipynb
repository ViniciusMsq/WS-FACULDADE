```python
import numpy as np
import cv2

#read image
imgEntrada = cv2.imread('bozoPalhaco.jpg')

#edge detection filter
mascara = np.array([[0.0, -1.0, 0.0], 
                   [-1.0, 4.0, -1.0],
                   [0.0, -1.0, 0.0]])

mascara = mascara/(np.sum(mascara) if np.sum(mascara)!=0 else 1)

hpassimg = cv2.filter2D(imgEntrada,-1,mascara)

#plt.subplot(),plt.imshow(imgEntrada),plt.title('Original')
plt.subplot(),plt.imshow(hpassimg),plt.title('passa alta')

```
