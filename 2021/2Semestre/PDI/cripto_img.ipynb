# Criptografia de imagem com Fernet

```python
from cryptography.fernet import Fernet

key = Fernet.generate_key()
fernet = Fernet(key)

print('chave privada: ',key.decode())
```

```python

with open('tetris.jpg', 'rb') as f:
    file = f.read()
    
img_criptografada = fernet.encrypt(file)


with open('tetris_enc.jpg', 'wb') as img_cr:
    img_cr.write(img_criptografada)

print('Arquivo criptografado')
```

```python
img_recuperada = fernet.decrypt(img_criptografada) 

data = img_recuperada
data = data.strip()

with open('tetris_dec.jpg', 'wb') as file:
    file.write(data)
    
    
```
