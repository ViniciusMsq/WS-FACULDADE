```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("corcel.jpg", cv2.COLOR_BGR2GRAY)

sift = cv2.xfeatures2d.SIFT_create()


keypoints, descriptors = sift.detectAndCompute(img, None)


img = cv2.drawKeypoints(img, keypoints, None)

cores = {"b": 0, "g": 1, "r": 2}

def histrgb(imagem):
    for cor in cores:
        histograma = cv2.calcHist([imagem],[cores[cor]],None,[256],[0,256])
        plt.plot(histograma,color = cor)
        plt.xlim([0,256])
    return plt.show()

def histh(img_hsv):
    img_hsv = cv2.cvtColor(img_hsv, cv2.COLOR_BGR2HSV)
    histograma = cv2.calcHist([img_hsv], [0], None, [180], [0, 180])
    plt.plot(histograma)
    return plt.show() 

histrgb(img)
histh(img)

plt.imshow(img)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,5))
plt.imshow(img_gray, cmap = "gray")



img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(10,5))
plt.imshow(img_hsv)


plt.figure(figsize=(10,5))
plt.imshow(img_hsv[:,:,0], cmap="gray")
plt.title("Matiz")

plt.figure(figsize=(10,5))
plt.imshow(img_hsv[:,:,1], cmap="gray")
plt.title("Saturação")

plt.figure(figsize=(10,5))
plt.imshow(img_hsv[:,:,2], cmap="gray")
plt.title("Brilho")



```

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt

img = cv2.imread('tetris.jpg') 
img2 = cv2.imread('corcel.jpg') 

img_1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
img_2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY) 

orb = cv2.ORB_create() 

keypoints1, Descriptors1 = orb.detectAndCompute(img_1,None) 
keypoints2, Descriptors2 = orb.detectAndCompute(img_2,None) 

matcher = cv2.BFMatcher() 
matches = matcher.match(Descriptors1,Descriptors2) 

img_result = cv2.drawMatches(img, keypoints1,img2, keypoints2, matches[:70],None) 


img_result = cv2.resize(img_result, (800,600)) 



cores = {"b": 0, "g": 1, "r": 2}

def histrgb(imagem):
    for cor in cores:
        histograma = cv2.calcHist([imagem],[cores[cor]],None,[256],[0,256])
        plt.plot(histograma,color = cor)
        plt.xlim([0,256])
    plt.show()


def histh(imagem_hsv):
    imagem_hsv = cv2.cvtColor(imagem_hsv, cv2.COLOR_BGR2HSV)
    histograma = cv2.calcHist([imagem_hsv], [0], None, [180], [0, 180])
    plt.plot(histograma)
    plt.show() 

histrgb(img2)
histh(img)
plt.imshow(img_result) 
```
